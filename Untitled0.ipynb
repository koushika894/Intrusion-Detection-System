# Network Intrusion Detection — Classic ML Pipeline
This notebook builds a binary classifier to identify anomalous network traffic using the `class` label. It demonstrates a clean, end‑to‑end workflow: data loading, exploration, preprocessing, feature selection, model training, hyperparameter tuning, and evaluation.

## Workflow
1. Load train/test datasets
2. Explore schema, missing values, and class balance
3. Encode categorical features and clean columns
4. Select top features with RFE + Random Forest
5. Standardize features and split train/validation
6. Train baseline models (Logistic Regression, KNN, Decision Tree)
7. Tune models with Optuna and evaluate precision/recall/F1

```python
def encode_train_test(train_df, test_df):
    # Handle the 'class' column in train_df separately, as it's the target and not in test_df
    if 'class' in train_df.columns and train_df['class'].dtype == 'object':
        label_encoder_class = LabelEncoder()
        train_df['class'] = label_encoder_class.fit_transform(train_df['class'])

    # Identify categorical columns common to both train and test (excluding 'class' if it was in train_df originally)
    common_object_cols = [col for col in train_df.columns if train_df[col].dtype == 'object' and col in test_df.columns]

    # Encode common categorical columns
    for col in common_object_cols:
        label_encoder = LabelEncoder()
        combined = pd.concat([train_df[col], test_df[col]], axis=0)
        label_encoder.fit(combined)
        train_df[col] = label_encoder.transform(train_df[col])
        test_df[col] = label_encoder.transform(test_df[col])

encode_train_test(train, test)
```

## 4. Data Preprocessing

### 4.1 Label Encoding
Convert all categorical features to numerical format using label encoding. We encode train and test together to ensure consistent mappings.

```python
train.drop(['num_outbound_cmds'], axis=1, inplace=True)
test.drop(['num_outbound_cmds'], axis=1, inplace=True)
```

### 4.2 Remove Constant Columns
Drop `num_outbound_cmds` as it has zero variance (constant value).

```python
train.head()
```

```python
X_train = train.drop(['class'], axis=1)
Y_train = train['class']
```

### 4.3 Feature Selection with RFE
Use Recursive Feature Elimination with Random Forest to select the top 10 most important features.

```python
rfc = RandomForestClassifier()

rfe = RFE(rfc, n_features_to_select=10)
rfe = rfe.fit(X_train, Y_train)

feature_map = [(i, v) for i, v in itertools.zip_longest(rfe.get_support(), X_train.columns)]
selected_features = [v for i, v in feature_map if i==True]

selected_features
```

```python
X_train = X_train[selected_features]
test = test[selected_features]
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
test = scale.transform(test)
```

### 4.4 Feature Scaling
Standardize features to have mean=0 and std=1 for optimal model performance.

```python
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, train_size=0.70, random_state=2)
```

### 5.2 K-Nearest Neighbors (KNN) with Optuna Tuning
Install Optuna for hyperparameter optimization.

```python
import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)
```

```python
def objective(trial):
    n_neighbors = trial.suggest_int('KNN_n_neighbors', 2, 16, log=False)
    classifier_obj = KNeighborsClassifier(n_neighbors=n_neighbors)
    classifier_obj.fit(x_train, y_train)
    accuracy = classifier_obj.score(x_test, y_test)
    return accuracy
```

```python
study_KNN = optuna.create_study(direction='maximize')
study_KNN.optimize(objective, n_trials=1)
print(study_KNN.best_trial)
```

```python
KNN_model = KNeighborsClassifier(n_neighbors=study_KNN.best_trial.params['KNN_n_neighbors'])
KNN_model.fit(x_train, y_train)

KNN_train, KNN_test = KNN_model.score(x_train, y_train), KNN_model.score(x_test, y_test)

print(f"Train Score: {KNN_train}")
print(f"Test Score: {KNN_test}")
```

```python
from sklearn.tree import DecisionTreeClassifier

clfd = DecisionTreeClassifier(criterion ="entropy", max_depth = 4)
start_time = time.time()
clfd.fit(x_train, y_train.values.ravel())
end_time = time.time()
print("Training time: ", end_time-start_time)
```

### 5.3 Decision Tree with Optuna Tuning

```python
start_time = time.time()
y_test_pred = clfd.predict(x_test)
end_time = time.time()
print("Testing time: ", end_time-start_time)
```

```python
def objective(trial):
    dt_max_depth = trial.suggest_int('dt_max_depth', 2, 32, log=False)
    dt_max_features = trial.suggest_int('dt_max_features', 2, 10, log=False)
    classifier_obj = DecisionTreeClassifier(max_features = dt_max_features, max_depth = dt_max_depth)
    classifier_obj.fit(x_train, y_train)
    accuracy = classifier_obj.score(x_test, y_test)
    return accuracy
```

```python
study_dt = optuna.create_study(direction='maximize')
study_dt.optimize(objective, n_trials=30)
print(study_dt.best_trial)
```

```python
dt = DecisionTreeClassifier(max_features = study_dt.best_trial.params['dt_max_features'], max_depth = study_dt.best_trial.params['dt_max_depth'])
dt.fit(x_train, y_train)

dt_train, dt_test = dt.score(x_train, y_train), dt.score(x_test, y_test)

print(f"Train Score: {dt_train}")
print(f"Test Score: {dt_test}")
```

## 6. Cross-Validation & Detailed Metrics

### 6.1 Setup Models for Cross-Validation

```python
from sklearn.model_selection import cross_val_score
models = {}
models['KNeighborsClassifier']= knn
models['LogisticRegression']= lr
models['DecisionTreeClassifier']= dtc
```

```python
scores = {}
for name in models:
  scores[name]={}
  for scorer in ['precision','recall']:
    scores[name][scorer] = cross_val_score(models[name], x_train, y_train, cv=10, scoring=scorer)
```

### 6.2 10-Fold Cross-Validation
Perform 10-fold cross-validation to evaluate precision and recall.

```python
def line(name):
  return '*'*(25-len(name)//2)

for name in models:
  print(line(name), name, 'Model Validation', line(name))

  for scorer in ['precision','recall']:
    mean = round(np.mean(scores[name][scorer])*100,2)
    stdev = round(np.std(scores[name][scorer])*100,2)
    print ("Mean {}:".format(scorer),"\n", mean,"%", "+-",stdev)
    print()
```

### 6.3 Print Cross-Validation Results

```python
for name in models:
    for scorer in ['precision','recall']:
        scores[name][scorer] = scores[name][scorer].mean()
scores=pd.DataFrame(scores).swapaxes("index", "columns")*100
scores.plot(kind = "bar",  ylim=[80,100], figsize=(24,6), rot=0)
```

### 6.4 Visualize Precision & Recall

```python
models = {}
models['KNeighborsClassifier']= knn
models['LogisticRegression']= lr
models['DecisionTreeClassifier']= dtc
```

## 7. Final Model Testing

### 7.1 Train Final Models

```python

preds={}
for name in models:
    models[name].fit(x_train, y_train)
    preds[name] = models[name].predict(x_test)
print("Predictions complete.")
```

### 7.2 Generate Predictions

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score
def line(name,sym="*"):
    return sym*(25-len(name)//2)
target_names=["normal","anomaly"]
for name in models:
    print(line(name), name, 'Model Testing', line(name))
    print(confusion_matrix(y_test, preds[name]))
    print(line(name,'-'))
    print(classification_report(y_test, preds[name], target_names=target_names))
```

### 7.3 Confusion Matrix & Classification Report

```python
f1s = {}
for name in models:
    f1s[name]=f1_score(y_test, preds[name])
f1s=pd.DataFrame(f1s.values(),index=f1s.keys(),columns=["F1-score"])*100
f1s.plot(kind = "bar",  ylim=[80,100], figsize=(10,6), rot=0)
```
